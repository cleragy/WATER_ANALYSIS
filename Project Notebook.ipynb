## 5. Exploratory Data Analysis (EDA) <a class="anchor" id="chapter5"></a>


We aimed to gain deeper insights into the variables and their interrelationships by conducting an **Exploratory Data Analysis (EDA)** on our dataset. The primary objectives included examining and summarizing the key attributes of the dataframe through data visualization techniques and statistical methods. Additionally, we explored the salient features present in the data, leveraging both graphical representations and quantitative analyses. This comprehensive exploration enabled us to uncover meaningful patterns and trends that can inform subsequent research endeavors and facilitate data-driven decision-making processes.

```python
# Libraries for data loading, manipulation and analysis
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Displays output inline
%matplotlib inline

# Libraries for Handling Errors
import warnings
warnings.filterwarnings('ignore')
# Load the dataset
df = pd.read_csv('fixed_file.csv')
# The copy of the dataframe
df_copy = df.copy()
# Convert the 'Date (DD/MM/YYYY)' column to datetime format with the correct date format
df['Date (DD/MM/YYYY)'] = pd.to_datetime(df['Date (DD/MM/YYYY)'], format='%d/%m/%Y')

# Group by date and calculate mean pH
daily_pH = df.groupby('Date (DD/MM/YYYY)')['pH'].mean()

# Plot pH levels over time
plt.figure(figsize=(10, 6))
plt.plot(daily_pH.index, daily_pH.values, marker='o', linestyle='-')
plt.title('Mean pH Levels Over Time')
plt.xlabel('Date')
plt.ylabel('Mean pH')
plt.grid(True)
plt.show()
```

**Results** :

The graph shows a fluctuating pattern in the mean pH levels over the given time period. In May 2023, the mean pH level starts around 8.3, then it decreases and exhibits some ups and downs until July 2023. From July to August 2023, there is a noticeable increase in the mean pH level, reaching a peak around 8.4.

After August 2023, the mean pH levels start to decline gradually, with some minor fluctuations. In September and October 2023, there is a more significant drop in the mean pH levels, reaching a low point around 7.7 to 7.8.

The fluctuations in the mean pH levels could be influenced by various factors such as changes in environmental conditions, industrial or agricultural activities, weather patterns, or natural processes. Without additional context or information, it is difficult to pinpoint the specific reasons behind these pH level variations.

Overall, this graph provides a visual representation of the temporal changes in mean pH levels, which can be useful for monitoring environmental conditions, identifying potential issues, or analyzing the impact of various factors on pH levels over time.

**Distribution of Dissolved Oxygen (DO) Levels** :

Generate a code to visually compare how Dissolved Oxygen (DO) Levels are distributed . This visualization helps to highlight the frequency and prevalence Dissolved Oxygen (DO) Levels , providing a clear comparison and deeper insight into our analysis.
 The code will also provide a Summary Statistics for Dissolved Oxygen (DO) Levels.

```python
# Check if 'Sample temperature (°C)' column exists
if 'Sample temperature (°C)' in df.columns:
    # Plot histogram of water temperature
    plt.figure(figsize=(8, 6))
    plt.hist(df['Sample temperature (°C)'], bins=20, color='skyblue', edgecolor='black')
    plt.title('Distribution of Water Temperature')
    plt.xlabel('Water Temperature (°C)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    # Summary statistics for water temperature
    print("Summary Statistics for Water Temperature:")
    print(df['Sample temperature (°C)'].describe())

    # Plot water temperature over time if 'Date (DD/MM/YYYY)' column exists
    if 'Date (DD/MM/YYYY)' in df.columns:
        # Convert the date column to datetime format
        df['Date (DD/MM/YYYY)'] = pd.to_datetime(df['Date (DD/MM/YYYY)'], format='%d/%m/%Y')
        # Group by date and calculate mean water temperature
        daily_temp = df.groupby('Date (DD/MM/YYYY)')['Sample temperature (°C)'].mean()

        # Plot mean water temperature over time
        plt.figure(figsize=(10, 6))
        plt.plot(daily_temp.index, daily_temp.values, marker='o', linestyle='-')
        plt.title('Mean Water Temperature Over Time')
        plt.xlabel('Date')
        plt.ylabel('Mean Water Temperature (°C)')
        plt.grid(True)
        plt.show()
    else:
        print("Date column not found in the dataset.")
else:
    print("Sample temperature column not found in the dataset.")
```

**Results** :

The histogram reveals that the majority of the DO measurements are concentrated around lower values, with a prominent peak between 0 and 2 mg/L. This suggests that a significant portion of the water samples had relatively low levels of dissolved oxygen.
However, there are also some observations at higher DO levels, as indicated by the smaller bars extending up to around 8 mg/L.

Based on the histogram and summary statistics, it can be inferred that the majority of the water samples had relatively low levels of dissolved oxygen, with a significant portion falling below the median value of 1.87 mg/L. However, there were also some observations with higher DO levels, potentially indicating variations in water quality or environmental conditions across the sampling locations or times.

**Mean dissolved oxygen over time**

The plot shows significant fluctuations in the mean DO levels throughout the observed period. There are several peaks and troughs, indicating variations in the levels of dissolved oxygen in the water over time.

In the initial months of May and June 2023, the mean DO levels are relatively high, with a peak around 4.5 mg/L in early June. However, the levels start to decline towards the end of June and remain low throughout July and August, with the lowest point occurring in late August around 2 mg/L.

From September onwards, the mean DO levels begin to increase again, reaching another peak in early October around 4.5 mg/L. This is followed by a gradual decrease until late November, when the levels seem to stabilize around 3 mg/L.

The fluctuating pattern suggests that the dissolved oxygen levels in the water are influenced by various factors that may vary over time. These factors could include temperature changes, biological activity, water flow rates, or other environmental conditions that affect the oxygen exchange between the water and the surrounding atmosphere.

It's important to note that dissolved oxygen is a crucial parameter for aquatic life, as many organisms, such as fish and other marine creatures, depend on adequate oxygen levels for respiration and survival. Sustained low levels of dissolved oxygen can have detrimental effects on the aquatic ecosystem.

Overall, this graph provides insights into the temporal variations in mean dissolved oxygen levels, highlighting periods of high and low concentrations. Such information can be valuable for monitoring water quality, identifying potential issues, and implementing appropriate measures to maintain healthy dissolved oxygen levels in the water body.
